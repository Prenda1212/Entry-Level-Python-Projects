```python
import pandas as pd
from pandas import Series, DataFrame

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic = pd.read_csv('train.csv')
```

```python
titanic
```

```python
titanic.columns=['PassengerId', 'Survived','Pclass',
                 'Name','Sex','Age', 'SibSp','Parch',
                 'Ticket','Fare','Cabin','Embarked']
```

```python
#Define a new item

def male_female_child(passenger):
      # Take the Age and Sex
    age,sex = passenger
    
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic['person'] = titanic[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
# Another way to visualize the data is to use FacetGrid to plot multiple kedplots on one plot

# Set the figure equal to a facetgrid with the pandas dataframe 
#as its data source, set the hue, and change the aspect ratio to 4.
fig = sns.FacetGrid(titanic, hue="Sex",aspect=4)

# Next use map to plot all the possible kdeplots 
#for the 'Age' column by the hue choice
#All this map will be mapped as the kde plots
fig.map(sns.kdeplot,'Age',shade= True)

# Set the x max limit by the oldest passenger
oldest = titanic['Age'].max()

#Since we know no one can be negative years old 
#set the x lower limit at 0
fig.set(xlim=(0,oldest))

#Finally add a legend
fig.add_legend()
```

```python
# Set the figure to the difference between child/male/female
fig = sns.FacetGrid(titanic, hue="person",aspect=4)

# Next use map to plot all the possible kdeplots 
#for the 'Age' column by the hue choice
#All this map will be mapped as the kde plots
#æ ¸å¯†åº¦å›¾kdeplot
fig.map(sns.kdeplot,'Age',shade= True)

# Set the x max limit by the oldest passenger
oldest = titanic['Age'].max()

#Since we know no one can be negative years old 
#set the x lower limit at 0
fig.set(xlim=(0,oldest))

#Finally add a legend
fig.add_legend()
```

```python
fig = sns.FacetGrid(titanic, hue="Pclass",aspect=4)

fig.map(sns.kdeplot,'Age',shade= True)

# Set the x max limit by the oldest passenger
oldest = titanic['Age'].max()

#Since we know no one can be negative years old 
#set the x lower limit at 0
fig.set(xlim=(0,oldest))

#Finally add a legend
fig.add_legend()
```

```python
# So let's grab that letter for the deck level with a simple for loop

# Set empty list
levels = []

#loop to grab first letter

for level in deck:
    levels.append(level[0])

cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']
sns.countplot('Cabin',data = cabin_df, palette='winter_d')
```

```python
# Redefine cabin_df as everything but where the row was equal to 'T'
cabin_df = cabin_df[cabin_df.Cabin != 'T']
#Replot
sns.countplot('Cabin',data=cabin_df,palette='summer')
```
