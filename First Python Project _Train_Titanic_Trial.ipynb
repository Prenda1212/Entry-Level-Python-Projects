```python
#Define 'more_child' based on sex and age

def more_child(psg): 
    age,sex = psg
    if age < 16:
        return 'child'
    else:
        return sex
    
#Create a new columns

titanic['person'] = titanic[['Age','Sex']].apply(more_child,axis=1)
```
